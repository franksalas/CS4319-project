```python
fig = sns.FacetGrid(uh_salary_df, hue='Sex', aspect=2)
fig.map(sns.kdeplot, 'AnnualSal', shade=True)

richest = uh_salary_df['AnnualSal'].max()
poorest = uh_salary_df['AnnualSal'].min()
fig.set(xlim=(300, richest))

fig.add_legend()
```

```python
fig = sns.FacetGrid(uh_salary_df, hue='DeptName', aspect=2)
fig.map(sns.kdeplot, 'AnnualSal', shade=True)

richest = uh_salary_df['AnnualSal'].max()
poorest = uh_salary_df['AnnualSal'].min()
middlest = uh_salary_df['AnnualSal'].mean()
fig.set(xlim=(0, richest))

fig.add_legend()
```
